# Task 1: Exploring and Visualizing a Simple Dataset

## Introduction


The objective of this project is to explore and visualize the Iris dataset. The Iris dataset is one of the most popular datasets in machine learning and contains measurements of different iris flower species. This project focuses on understanding the dataset structure, performing exploratory data analysis, and visualizing relationships between variables using different plots.


## Importing Useful Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = (8,5)
```

## Loading DataSets

```python
iris = sns.load_dataset("iris")
```

```python
iris
```

## DataSet Inspection

```python
print("Dataset Shape:")
print(iris.shape)

print("\nColumn Names:")
print(iris.columns)

print("\nFirst 5 Rows:")
print(iris.head())

print("\nDataset Information:")
print(iris.info())

print("\n Summary:")
print(iris.describe())

print("\nMissing Values:")
print(iris.isnull().sum())
```

## Exploratory Data Analysis

### Species Count

```python
print("\nSpecies Distribution:")
print(iris["species"].value_counts())
```

### Correlation Matrix

```python
print("\nCorrelation Matrix:")
print(iris.corr(numeric_only=True))
```

## Scatter Plot

```python
plt.figure(figsize=(8,6))

sns.scatterplot(
    data=iris,
    x="petal_length",
    y="petal_width",
    hue="species",
    s=100
)

plt.title("Petal Length vs Petal Width")
plt.show()
```

## Histogram

```python
plt.figure(figsize=(8,6))

sns.histplot(
    iris["sepal_length"],
    bins=15,
    kde=True
)

plt.title("Distribution of Sepal Length")
plt.xlabel("Sepal Length")
plt.show()
```

## Box Plot

```python
plt.figure(figsize=(8,6))

sns.boxplot(
    data=iris,
    x="species",
    y="sepal_length"
)

plt.title("Sepal Length by Species")
plt.show()
```

## Pairplot

```python
sns.pairplot(
    iris,
    hue="species"
)

plt.show()
```

## Heatmap

```python
plt.figure(figsize=(8,6))

sns.heatmap(
    iris.corr(numeric_only=True),
    annot=True,
    cmap="coolwarm"
)

plt.title("Correlation Heatmap")
plt.show()
```

## Conclusion

The Iris dataset was successfully loaded and explored using Python libraries such as Pandas, Matplotlib, and Seaborn. Various visualizations including scatter plots, histograms, and box plots were created to understand relationships, distributions, and outliers within the dataset. The analysis provided valuable insights into the characteristics of different iris species and demonstrated fundamental data exploration and visualization techniques.
